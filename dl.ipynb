# import dependencies

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import PillowWriter
from IPython.display import display, Image
```

Util functions

```python
def factorielle(n):
    """ cette fonction retourne le factorielle de n : n! """
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorielle(n-1)
```

Paramètres du développement limité

```python
x_value = 0
ordre_n = 5
```

# Développement limité de la fonction exp(x) au point x = 0

```python
def developpement_limite_exp(x, n):
    """ 
    Fonction qui calcul le développement limité de la fonction exponentielle 'exp(x)' au voisinage du point x=0
    """
    dl = 0
    for k in range(n):
        dl += pow(x,k)/factorielle(k)
    return dl
```

## Visualisation des deux courbes

```python
# Génération de points pour le tracé graphique
x_vals = np.linspace(0, 10, 100)
y_exp = np.exp(x_vals) 

fig = plt.figure()

l, = plt.plot([],[],"k--")
g, = plt.plot([],[],"r--")

plt.xlim(0,5)
plt.ylim(np.exp(0), np.exp(5))
plt.title('exp(x)')
writer = PillowWriter(fps=5)

l.set_data(x_vals, y_exp)

with writer.saving(fig, "expDL.gif", 100):
    for i in range(10):
        ylist = [developpement_limite_exp(xval,ordre_n+i) for xval in x_vals]
        
        g.set_data(x_vals, ylist)
        
        writer.grab_frame()
plt.close()

#  diplay gif animation 
display(Image(filename='expDL.gif'))
```

# Développement limité de la fonction trigonométrique 'cos(x)' au point x = 0

```python
def developpement_limite_cos(x, n):
    """ 
    Fonction qui calcul le développement limité de la fonction trigonométrique 'cos(x)' au voisinage du point x=0
    """
    dl = 0
    for k in range(int(np.floor(n/2))):
        dl += pow(-1,k)*(pow(x,2*k)/factorielle(2*k))
    return dl
```

## Visualisation des deux courbes

```python
x_vals = np.linspace(-5,5, 100)
y_cos = np.cos(x_vals) 

fig = plt.figure()

l, = plt.plot([],[],"k--")
g, = plt.plot([],[],"r--")

plt.xlim(-5,5)
plt.ylim(-1,1)

plt.title('cos(x)')
writer = PillowWriter(fps=5)

l.set_data(x_vals, y_cos)

with writer.saving(fig, "cosDL.gif", 100):
    for i in range(10):
        ylist = [developpement_limite_cos(xval,ordre_n+i) for xval in x_vals]
        
        g.set_data(x_vals, ylist)
        
        writer.grab_frame()
plt.close()

#  diplay gif animation 
display(Image(filename='cosDL.gif'))
```

# Affichage du DL de la fonction 'sin(x)' au point x = 0

```python
def developpement_limite_sin(x, n):
    """ 
    Fonction qui calcul le développement limité de la fonction trigonométrique 'sin(x)' au voisinage du point x=0
    """
    dl = 0
    for k in range(int(n)):
        dl += pow(-1,k)*(pow(x,2*k+1)/factorielle(2*k+1))
    return dl
```

## Visualisation des deux courbes

```python
x_vals = np.linspace(-5,5, 100)
y_sin = np.sin(x_vals) 

fig = plt.figure()

l, = plt.plot([],[],"k--")
g, = plt.plot([],[],"r--")

plt.xlim(-5,5)
plt.ylim(-1,1)

plt.title('sin(x)')
writer = PillowWriter(fps=5)

l.set_data(x_vals, y_sin)

with writer.saving(fig, "sinDL.gif", 100):
    for i in range(10):
        ylist = [developpement_limite_sin(xval,ordre_n+i) for xval in x_vals]
        
        g.set_data(x_vals, ylist)
        
        writer.grab_frame()
plt.close()

#  diplay gif animation 
display(Image(filename='sinDL.gif'))
```

# Affichage du DL de la fonction 'ln(x+1)' au point x = 0

```python
def developpement_limite_ln(x,n):
    """ 
    Fonction qui calcul le développement limité de la fonction logaritmique 'ln(x+1)' au voisinage du point x=0
    """
    dl = 0
    for k in range(n+2):
        if k != 0 :
            dl += pow(-1,k+1)*(pow(x,k)/k)
    return dl
```

## Visualisation des deux courbes

```python
x_vals = np.linspace(0,10, 100)
y_ln_vals = []
for v in x_vals:
    y_ln_vals.append(np.log(v+1))

fig = plt.figure()

l, = plt.plot([],[],"k--")
g, = plt.plot([],[],"r--")

plt.xlim(0,1)
plt.ylim(0,1)

plt.title("ln(x+1)")

writer = PillowWriter(fps=5)

l.set_data(x_vals, y_ln_vals)

with writer.saving(fig, "lnDL.gif", 100):
    for i in range(10):
        ylist = [developpement_limite_ln(xval,ordre_n+i) for xval in x_vals]
        
        g.set_data(x_vals, ylist)
        
        writer.grab_frame()
plt.close()

#  diplay gif animation 
display(Image(filename='lnDL.gif'))
```
